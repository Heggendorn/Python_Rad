Criando um bd em sqlite


```python
#importando a biblioteca
import sqlite3

#abrindo conexão
connection = sqlite3.connect("minhabase.db")

tabela = connection.cursor()  # cursor permite executar comandos sql

#se não existir o banco é criado automaticamente

tabela.execute("""
        CREATE TABLE IF NOT EXISTS  
               dados( nome TEXT, rg TEXT, cpf TEXT, data_nasc TEXT)
              """)

```

```python
#inserindo dados no banco
tabela.execute("""INSERT INTO dados VALUES
               ('Maria', '00235', '23457891', '14/02/1980')
               """)
#Salva os dados no banco
connection.commit()
```

```python
# inserindo dados multiplos

dados = [
          ('Joao','15156156','211211515','15/07/2010'),
          ('andre','1451561','15151651','20/03/2000'),
          ('maria','15165','1548748','11/12/1971'),
          ('pedro','994848','1515616','30/04/1979')
          ]

tabela.executemany("INSERT INTO dados Values (?,?,?,?)", dados)
connection.commit()
```

```python
print("\nLista de todos os registros na tabela: \n")
for i in tabela.execute("SELECT rowid, * FROM dados ORDER BY nome"):
    print(i)
```

```python
#deletando dados

sql_del = """DELETE FROM dados WHERE nome  = 'pedro' """
tabela.execute(sql_del)
connection.commit()
```
